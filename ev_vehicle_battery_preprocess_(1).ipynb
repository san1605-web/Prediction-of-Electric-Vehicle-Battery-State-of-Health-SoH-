null values in Car_Model,Driving_Style is filled with mode

```python
print(df["Car_Model"].mode())
df['Car_Model']=df['Car_Model'].fillna("BYD Atto 3")
df["Car_Model"].head(10)
```

```python
print(df["Driving_Style"].mode())
df["Driving_Style"].str.lower()
df["Driving_Style"].str.strip()
df["Driving_Style"]=df["Driving_Style"].fillna("Moderate")
df["Driving_Style"].head()
```

```python
df["Car_Model"]=df["Car_Model"].str.lower()
df["Car_Model"]=df["Car_Model"].str.strip()
```

```python
print(df["Car_Model"].unique())

df.loc[df["Battery_Type"].isnull()&(df["Car_Model"]=='wuling air ev'),"Battery_Type"]="LFP"
df.loc[df["Battery_Type"].isnull()&(df["Car_Model"]=='tesla model 3'),"Battery_Type"]="NMC"
df.loc[df["Battery_Type"].isnull()&(df["Car_Model"]=='ford mustang mach-e'),"Battery_Type"]="NMC"
df.loc[df["Battery_Type"].isnull()&(df["Car_Model"]=='hyundai ioniq 5'),"Battery_Type"]="NMC"
df.loc[df["Battery_Type"].isnull()&(df["Car_Model"]=='byd atto 3'),"Battery_Type"]="LFP"
```

null values in Battery_Capacity_kWh,Total_Charging_Cycles,Fast_Charge_Ratio,Internal_Resistance_Ohm is filled with median

```python
print(df["Battery_Capacity_kWh"].median())
df["Battery_Capacity_kWh"]=df["Battery_Capacity_kWh"].fillna(72)
df["Battery_Capacity_kWh"]=df["Battery_Capacity_kWh"].astype(int)
df["Battery_Capacity_kWh"].head(7)
```

```python
print(df["Total_Charging_Cycles"].median())
df["Total_Charging_Cycles"]=df["Total_Charging_Cycles"].fillna(df["Total_Charging_Cycles"].median())
df["Total_Charging_Cycles"].head(7)
```

```python
print(df["Fast_Charge_Ratio"].median())
df["Fast_Charge_Ratio"]=df["Fast_Charge_Ratio"].fillna(df["Fast_Charge_Ratio"].median())
df["Fast_Charge_Ratio"].head(200)
```

```python
print(df["Internal_Resistance_Ohm"].median())
df["Internal_Resistance_Ohm"]=df["Internal_Resistance_Ohm"].fillna(df["Internal_Resistance_Ohm"].median())
df["Internal_Resistance_Ohm"].head(10)
```

null values in Vehicle_Age_Months,Avg_Temperature_C,Avg_Discharge_Rate_C is filled with mean

```python
print(df["Vehicle_Age_Months"].mean())
df["Vehicle_Age_Months"]=df["Vehicle_Age_Months"].fillna(65)
df["Vehicle_Age_Months"].head(2100)
```

```python
print(df["Avg_Temperature_C"].mean())
df["Avg_Temperature_C"]=df["Avg_Temperature_C"].fillna(32.09)
df["Avg_Temperature_C"].head(10)
```

```python
df["Battery_Status"]=df["Battery_Status"].str.lower()
df["Driving_Style"]=df["Driving_Style"].str.lower()
```

```python
print(df["Avg_Discharge_Rate_C"].mean())
df["Avg_Discharge_Rate_C"]=df["Avg_Discharge_Rate_C"].fillna(2.00)
df["Avg_Discharge_Rate_C"].head(10)
```
